Title: Data Inconsistencies

Types of Data inconsistencies.
  1. Inconsistent Formats
  2. Inconsistent Naming conventions
  3. Typographical error
  4. Duplicates
  5. Contradictory
 

```python
#  libraries
import pandas as pd
```

```python
# Data Frame
data = {
    'date': ['2021-12-1', '01-12-2022', '2022/12/01', '12-01-2021'],
    'country': ['USA', 'U.S.A', 'America', 'United States'],
    'name': ['Jhon Doe', 'Jonh Doe', 'Jane Doe', 'Jane Deo'],
    'sales_2020': [100, 200, None, 200],
    'sales_2021': [None, 150, 300, 150]
}

# make pandas data frame
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# standardizing the date format
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date'] = df['date'].dt.strftime('%Y-%m-%d')
```

```python
df.head()
```

```python
df['date'] = df['date'].ffill()
```

```python
df.head()
```

```python
# harmonize the name of the country
country_mapping = {'USA': 'United States', 'U.S.A': 'United States', 'America': 'United States'}
df['country'] = df['country'].replace(country_mapping)
```

```python
df.head()
```

```python
# Corect the typographical mistake in name
df['name'] = df['name'].replace({'Jonh Doe': 'Jhon Doe', 'Jane Deo': 'Jane Doe'})
df.head()
```

```python
# Remove duplicates
df = df.drop_duplicates(subset='name')
df.head()
```

```python
df.head()
```

```python
# Resolving Contradictory Data
# For demenstration, Let assume sales 2021 are more than sales 2020
# We will remove the rows where these conditions are not met
df = df.drop(df[df['sales_2021'] <= df['sales_2020']].index)
df.head()
```
